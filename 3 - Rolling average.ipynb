# Part 3 - Rolling Average

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('ZnO.xye')

def fig_setup(figsize=(10, 10)):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111)
    return fig, ax
```

Function rolling_average returns two arrays: 
- new_x = the corresponding x values to those in moving_average
- moving_average = the rolling average of the data over a window n

An offset has been included to more easily see the difference various n values make.

```python
def rolling_average(data, n, offset=0.15):
    moving_average = []
    new_x = []
    for i in range(0, (len(data[:,0]) - n + 1)):
        window = data[:,1][i : i + n]
        window_average = sum(window) / n
        moving_average.append(window_average)
        new_x.append(data[:,0][int(i+(n/2))] + (offset*(n)))
    return (new_x, moving_average)
```

```python
fig, ax = fig_setup()
plt.plot(data[:,0], data[:,1], alpha = 0.3)
plt.plot(rolling_average(data, 2)[0], rolling_average(data, 2)[1], alpha = 0.2)
plt.plot(rolling_average(data, 5)[0], rolling_average(data, 5)[1], alpha = 0.2)
plt.legend(['Data', 'n=2', 'n=3'])
plt.xlabel('2θ')
plt.ylabel('Intensity')
plt.title('Diffraction Pattern')
plt.show()
plt.clf()
```
