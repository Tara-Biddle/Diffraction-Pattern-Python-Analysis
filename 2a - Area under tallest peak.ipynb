```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('ZnO.xye')

def fig_setup(figsize=(10, 10)):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111)
    return fig, ax
```

```python
# finds the maximum peak and the area under said peak within a certain width. 
def get_max_area(data, width=1.5):
    max_index = np.where(data[:,1] == np.max(data[:,1]))
    xmin = data[:,0][max_index] - (width/2)
    xmax = data[:,0][max_index] + (width/2)
    index_xmin = np.searchsorted(data[:,0], xmin)
    index_xmax = np.searchsorted(data[:,0], xmax)
    plt.plot(data[:,0], data[:,1])
    ax.set_xlim(xmin, xmax)
    plt.xlabel('2Î¸')
    plt.ylabel('Intensity')
    plt.title('Diffraction Pattern of Tallest Peak')
    plt.show()
    plt.clf()
    
    x = data[:,0][index_xmin[0]:(index_xmax[0] + 1)]
    y = data[:,1][index_xmin[0]:(index_xmax[0] + 1)]
    area = np.trapz(y, x, dx = 0.1)
    print("area under tallest peak =", area)
    
fig, ax = fig_setup()    
get_max_area(data)
```
