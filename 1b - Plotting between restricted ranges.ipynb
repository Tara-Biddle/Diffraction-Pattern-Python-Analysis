```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('ZnO.xye')
```

```python
def fig_setup(figsize=(10, 10)):
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111)
    return fig, ax

# plots between a specified x range
def plot_range(data, xmin = np.min(data[:,0]), xmax = np.max(data[:,0])):
    plt.plot(data[:,0], data[:,1])
    plt.xlabel('2θ')
    plt.ylabel('Intensity')
    plt.title('Diffraction Pattern Between ' + str(xmin) + ' and ' + str(xmax))
    ax.set_xlim([xmin, xmax])
    plt.show()
    plt.clf()
    
fig, ax = fig_setup()    
plot_range(data)

fig, ax = fig_setup()
plot_range(data, xmin = 31, xmax = 33)
```

```python
# plots between specified x range and scales y axis of plot appropriately, 
# and returns minimum and maximum y values in specified range
def plot_range_scale(data, xmin = np.min(data[:,0]), xmax = np.max(data[:,0])):
    min_index = np.searchsorted(data[:,0], xmin)
    max_index = np.searchsorted(data[:,0], xmax)
    y_range = data[:,1][min_index:(max_index + 1)]
    plt.plot(data[:,0], data[:,1])
    plt.xlabel('2θ')
    plt.ylabel('Intensity')
    plt.title('Diffraction Pattern Between ' + str(xmin) + ' and ' + str(xmax))
    ax.set_xlim([xmin, xmax])
    ax.set_ylim([np.min(y_range) - 10, np.max(y_range) + 10])
    plt.show()
    plt.clf()
    print('Min y is ' + str(np.min(y_range)))
    print('Max y is ' + str(np.max(y_range)))
              
fig, ax = fig_setup()    
plot_range_scale(data, 31, 33)
```

```python
# xmin and xmax default to None
def plot_range_default(data, xmin=None, xmax=None):
    if type(xmin) == type(None):
        plt.plot(data[:,0], data[:,1])
        plt.xlabel('2θ')
        plt.ylabel('Intensity')
        plt.title('Diffraction Pattern')
        ax.set_xlim([np.min(data[:,0]), np.max(data[:,0])])
        plt.show()
        plt.clf()
    else:
        min_index = np.searchsorted(data[:,0], xmin)
        max_index = np.searchsorted(data[:,0], xmax)
        y_range = data[:,1][min_index:(max_index + 1)]
        plt.plot(data[:,0], data[:,1])
        plt.xlabel('2θ')
        plt.ylabel('Intensity')
        plt.title('Diffraction Pattern Between ' + str(xmin) + ' and ' + str(xmax))
        ax.set_xlim([xmin, xmax])
        ax.set_ylim([np.min(y_range) - 10, np.max(y_range) + 10])
        plt.show()
        plt.clf()
              
fig, ax = fig_setup()    
plot_range_default(data)

fig, ax = fig_setup()    
plot_range_default(data, xmin=31, xmax=33)
```

```python
# plots over an x range of specified width with the largest peak in the diffraction pattern at the centre
def plot_max(data, width=1.5):
    max_index = np.where(data[:,1] == np.max(data[:,1]))
    plt.plot(data[:,0], data[:,1])
    ax.set_xlim([data[:,0][max_index] - (width/2), data[:,0][max_index] + (width/2)])
    plt.xlabel('2θ')
    plt.ylabel('Intensity')
    plt.title('Diffraction Pattern of Tallest Peak')
    plt.show()
    plt.clf()
    
fig, ax = fig_setup()    
plot_max(data)
```
